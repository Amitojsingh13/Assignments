# Question 1


```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
norm = st.norm(1.5,0.5)
draws = norm.rvs(1000)
```

```python
sns.kdeplot(draws)
```

```python
mean = np.mean(draws)
var = np.var(draws)
skew = st.skew(draws)
kur = st.kurtosis(draws)
print("Sample mean is:",mean)
print("Variance is: ",var)
print("Skewness is: ",skew)
print("Kurtosis is: ",kur)
```

```python
Mad = st.median_abs_deviation(draws)
std = 1.482 * Mad
print("Standard Deviation using MAD: ",std)
```

```python
q1,q3 = np.percentile(draws,(27,75))
Iqr = q3 - q1
SigmaG = 0.7613*Iqr
print("SigmaG of samples are: ",SigmaG)
```

# Question 2


```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
points = np.linspace(-10, 10, 1000)
cauch = st.cauchy(0, 1.5)
gaus = st.norm(0,1.5)
```

```python
plt.plot(points,cauch.pdf(points))
plt.plot(points,gaus.pdf(points),'--')
plt.legend(['Cauchy', 'Gaussian'])
```

# Question 3


```python
import  numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
arr = np.arange(-20,20,1)
gaus = st.norm(5,np.sqrt(5))
pois = st.poisson(5)
```

```python
plt.plot(arr,gaus.pdf(arr))
plt.plot(arr,pois.pmf(arr),'--')
plt.legend(['Gaussian','Poisson'])
```
